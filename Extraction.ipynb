# Rijks Museum
---------------

```python
url = f'https://www.rijksmuseum.nl/api/nl/collection/SK-C-5?key={rijks_api_key}'
r = requests.get(url).json()
```

```python
# Define the base URL for the Rijksmuseum API
base_url = "https://www.rijksmuseum.nl/api/en"

# Set the page size (number of items per page)
page_size = 100 

# Initialize variables for pagination
page_number = 1
all_items = []

while True:
    url = f"{base_url}/collection?key={rijks_api_key}&p={page_number}&ps={page_size}"
    response = requests.get(url)
    
    if response.status_code == 200:
        page_data = response.json()
        
        # Extracting objects
        items_on_page = page_data["artObjects"]
        all_items.extend(items_on_page)

        # Check if there are more pages to retrieve
        if len(items_on_page) < page_size:
            break  # All items have been retrieved
        else:
            page_number += 1  # Move to the next page
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
        break

# Now, all_items contains information about all items in the collection
print(f"Total items retrieved: {len(all_items)}")
```

```python
rijks_df = pd.json_normalize(all_items)
rijks_df.to_csv(r'C:\Users\USER\Desktop\Learning\ArtifactAnalysis\Resources\rijksmuseum.csv')
```
